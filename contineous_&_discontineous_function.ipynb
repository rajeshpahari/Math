```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
#contineous function:

xx=np.linspace(-5,1,1000)
y=2**xx

plt.plot(xx,y,label='contineous function')
```

```python
#Discontineous function:Jumped discontinuty
y1= (xx/2)       * (xx<0)
y2=((xx**2)+1)   * ((xx>=0)&(xx<1))



plt.plot(xx[xx<0],.5*xx[(xx<0)],label='fun1')
plt.plot(xx[(xx>=0)&(xx<1)],1+xx[(xx>=0)&(xx<1)]**2,label='Fun2')
plt.title('Jumped discontinuty')
```

```python
#infinite discontineuity
xx=np.linspace(-10,10,100)
plt.ylim([-2,2])
plt.plot(xx,np.tan(xx))
```

```python
#Removable discontineuty
xx=np.linspace(0,6,100)
plt.plot(xx,(xx**2-2**xx)/(xx**2-4));
```

```python
xx=np.arange(-1,2,.1)
#list of functions
y=[0]*3
y[0]=np.sin(xx*np.pi)
y[1]=1.5 * np.ones(len(xx))
y[2]=-(xx-2)**2

#domain
xdomains=[0]*3
xdomains[0]=xx<0
xdomains[1]=np.abs(xx)<.1/2  #as there is no zero value in xx, so chosing nearest to zero
xdomains[2]=xx>0

marker='-o-'
for i in range(len(y)):
  plt.plot(xx[xdomains[i]],y[i][xdomains[i]],marker[i])


plt.show()
```

```python
from sympy.abc import x
fx=sym.Piecewise(
    (sym.sin(x*sym.pi),(x<0)),
    (1.5,sym.Eq(x,0)),
    ((-(x-2)**2),(x>0))
)


sym.plot(fx,(x,xx[0],xx[-1]))
```

```python
#removable discontinuty
y=np.sin(xx*np.pi)+xx**2
y[np.argmin(np.abs(xx-0))]=np.pi
plt.plot(xx,y,'o')
```
