```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

```

```python
xx=np.linspace(-3,5,100)
fun1=0            *   (xx<0)
fun2=-2*xx        *   ((xx>=0)&(xx<3))
fun3=(xx**3)*.1   *   (xx>=3)

y=fun1+fun2+fun3

plt.plot(xx,y)
    
```

```python
#Piecewise function plot
xx=np.linspace(-3,5,1234)
fun1=0            *   (xx<0)
fun2=-2*xx        *   ((xx>=0)&(xx<3))
fun3=(xx**3)*.1   *   (xx>=3)

y=fun1+fun2+fun3

plt.plot(xx[xx<0],np.zeros(len(xx[xx<0])),label='fun1')
plt.plot(xx[((xx>=0)&(xx<3))],-2*xx[((xx>=0)&(xx<3))],label='fun2')
plt.plot(xx[xx>=3],xx[xx>=3]**3/10,label='fun3')
plt.legend()
plt.show()
```

```python
from sympy.abc import x
fun1=0   
fun2=-2*x
fun3=(x**3)*.1 
fx=sym.Piecewise( (fun1, x<0) ,(fun2,(x>=0)&(x<3)), (fun3,x>=3) )
fx
sym.plot(fx,(x,-3,5))
```

```python
#find out the value of y in nearest point of .5
loc=np.argmin(abs(xx-.5))
y[loc]
```
