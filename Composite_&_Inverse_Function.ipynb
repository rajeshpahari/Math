```python

import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
#Composite function using numpy
xx=np.linspace(-4,4,1000)

def fx(xx):
  return xx**2-3

def gx(xx):
  return 7*np.abs(xx)+3

fgx=fx(gx(xx))
gfx=gx(fx(xx))

plt.plot(xx,fx(xx),label='fx')
plt.plot(xx,gx(xx),label='gx')

plt.plot(xx,fgx,label='fgx')
plt.plot(xx,gfx,label='gfx')


plt.legend()
plt.ylim([-10,50])
plt.show()
```

```python
xx=np.linspace(-100*np.pi,100*np.pi,1001)
def fx(xx):
  return np.sin(xx)

def gx(xx):
  return np.log(xx)

def hx(xx):
  return 2*xx**2+5

f3=fx(gx(hx(xx)))

plt.plot(xx,f3)
```

```python
#Inverse function

xx=np.linspace(.1,5,1001)
def fx(xx):
  return np.log(2*xx)

def gx(xx):
  return np.exp(xx)/2



fgx=fx(gx(xx))
gfx=gx(fx(xx))

plt.plot(xx,fx(xx),label='fx')
plt.plot(xx,gx(xx),label='gx')
plt.plot(xx,gfx,label='gfx')
plt.plot(xx,fgx,'.',label='fgx')
plt.ylim([-5,10])
plt.legend()
plt.show()
```

```python
#Composite function using sympy
sx=sym.symbols('x')

fx=sym.sin(sx)
gx=sym.log(sx)
hx=2*sx**2+5

fx.subs(sx,gx.subs(sx,hx))

```

```python
sx=sym.symbols('x')

fx=sym.sin(sx)
gx=sym.log(sx)
hx=2*sx**2+5

f3=fx.subs(sx,gx.subs(sx,hx))


sym.plot(f3,(sx,-100,100),title='f3')
```

```python
#invert function in sympy
y=sym.symbols('y')
f=2*sx+3
f
```
