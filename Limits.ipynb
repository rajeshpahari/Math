```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def fx(u):
  return np.cos(u**2)**2+np.pi

xx=np.linspace(-2.1,2.1,211)
plt.plot(xx,fx(xx))
plt.grid()
plt.show()
```

```python
# Find out the limit of above function at x=1

#define target point
a=1

#Starting poistion we will take a-1 to start from left and a+1 from right

x0=np.array([a-1,a+1])


#Initializations
iterations=10
xaxisvals=np.zeros((iterations,2))
limitvals=np.zeros((iterations,2))


#Evaluate and store x0 & f(x0) 

for i in range(iterations):
  xaxisvals[i,:]=x0
  limitvals[i,:]=fx(x0)

  #Update x0
  x0=(x0+a)/2


```

```python
print('Limits from left:')
print(np.vstack((xaxisvals[:,0],limitvals[:,0])).T)

print('Limits from right:')
print(np.vstack((xaxisvals[:,1],limitvals[:,1])).T)

print('Value of fx at x0=a:')
print(fx(a))
```

```python
#Visualise the result

plt.plot(xx,fx(xx),'k')
plt.plot([a,a],[np.pi,1+np.pi],'k--',lw=.2)
plt.plot(xaxisvals,limitvals,'o',markerfacecolor='w')
plt.title(f'f({a})={fx(a)}')
plt.show()
```

```python
#Another Limit using sympy
def fx(x):
  return np.cos(x*np.pi)+np.log(x)**2
```

```python
xx=np.linspace(0,5,500)
plt.grid()
plt.plot(xx,fx(xx));
```

```python
expfact=4
g=1.0000001-np.linspace(1,.00001**expfact,10)**(1/expfact)
xfromleft=3-g
xfromright=3+g

#function values
limitfromleft=fx(xfromleft)
limitfromright=fx(xfromright)


print(f'Limit approchaes {limitfromleft[0]} from the left')
print(f'Limit approchaes {limitfromright[0]} from the right')
print(f'Limit is         {fx(3)} at 3')


plt.plot(xx,fx(xx),label='f(x)')
plt.plot(xfromleft,limitfromleft,'s',label='From left')
plt.plot(xfromleft,limitfromright,'o',label='From right')
plt.legend()
plt.show();
```

```python

from sympy.abc import x
sfx=sym.cos(x*sym.pi)+sym.log(x)**2
sfx
print('correct answer:',sfx.subs(x,3))
print('incorrect answer:',fx(3))
print('correct answer:  ',sym.N(sfx.subs(x,3)))
```

```python
sym.limit(sfx,x,3 )
```

```python
sym.limit(sfx,x,3 ,dir='+')
```

```python
#infinite Limits
def fx(x):
  return 1/(x-2)**2
```

```python
xx=np.linspace(0,4,1000)
plt.ylim([0,100])
plt.plot(xx,fx(xx))
```

```python
#Limit at 2
expfact=4
g=1.0000001-np.linspace(1,.00001**expfact,10)**(1/expfact)
xfromleft=2-g
xfromright=2+g

limitfromleft=fx(xfromleft)
limitfromright=fx(xfromright)
print(f'Limit approaches from left:  {limitfromleft[0]}')
print(f'Limit approaches from right: {limitfromright[0]}')
#print(f'Limit at 2: {fx(2)}')

plt.plot(xfromleft,limitfromleft,'s',label='limit from left')
plt.plot(xfromright,limitfromright,'s',label='limit from right')
#plt.plot(2,fx(2),'s',label='limit at 2')
plt.legend()
plt.ylim([-2,2])
plt.show()
```

```python
from sympy.abc import x
sfx=1/(x-2)**2
sfx
sfx_lam=sym.lambdify(x,sfx)
print('correct answer:',sfx.subs(x,2))
#print('incorrect answer:',fx(2))
print('correct answer:  ',sym.N(sfx.subs(x,2)))

plt.plot(xx,sfx_lam(xx))
```
