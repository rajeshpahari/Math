```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fx(u):
  return np.cos(u**2)**2+np.pi

xx=np.linspace(-2.1,2.1,211)
plt.plot(xx,fx(xx))
plt.grid()
plt.show()
```

```python
# Find out the limit of above function at x=1

#define target point
a=1

#Starting poistion we will take a-1 to start from left and a+1 from right

x0=np.array([a-1,a+1])


#Initializations
iterations=10
xaxisvals=np.zeros((iterations,2))
limitvals=np.zeros((iterations,2))


#Evaluate and store x0 & f(x0) 

for i in range(iterations):
  xaxisvals[i,:]=x0
  limitvals[i,:]=fx(x0)

  #Update x0
  x0=(x0+a)/2


```

```python
print('Limits from left:')
print(np.vstack((xaxisvals[:,0],limitvals[:,0])).T)

print('Limits from right:')
print(np.vstack((xaxisvals[:,1],limitvals[:,1])).T)

print('Value of fx at x0=a:')
print(fx(a))
```

```python
#Visualise the result

plt.plot(xx,fx(xx),'k')
plt.plot([a,a],[np.pi,1+np.pi],'k--',lw=.2)
plt.plot(xaxisvals,limitvals,'o',markerfacecolor='w')
plt.title(f'f({a})={fx(a)}')
plt.show()
```
