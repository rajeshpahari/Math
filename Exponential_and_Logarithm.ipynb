```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

from IPython import display
```

```python
n=[1,2,5,10]



for i in n:
  e=(1+(1/i))**i
  print (f'input: {i:2.0f} Estimate:{e:.5f}   Difference to e from estimate:{(np.exp(e)-e):.5f}')

```

```python
#draw Graph of the above differences for 1000 instances..observe how the plot is converging
n=np.arange(1,1001)
ediff=np.exp(1)-(1+(1/n))**n
plt.plot(n,ediff);
```

```python
#Some functions

plt.figure(figsize=(10,10))
x=np.linspace(-2,2,53)
y1=np.exp(x)
plt.plot(x,y1,label='$y=e^x$')

y2=np.exp(x**2)
plt.plot(x,y2,label='$y=e^{x^2}$')

y3=np.exp((-x)**2)
plt.plot(x,y3,label='$y=e^{(-x)^2}$')

y4=np.exp(-(x**2))
plt.plot(x,y4,label='$y=e^{-(x^2)}$')


y5=np.exp(x)**2
plt.plot(x,y5,label='$y=(e^x)^2$')
plt.legend()
plt.ylim([-1,10])
```

```python
#Show that natural log and natural exp are inverse of each other
##Look into the warning message - what does it mean??
x=np.linspace(-2,2,53)
y1=np.log(x)
y2=np.exp(x)
y3=np.log(np.exp(x))
y4=np.exp(np.log(x))

plt.plot(x,y1,label='$y=log(x)$')
plt.plot(x,y2,label='$y=(e^x)$')
plt.plot(x,y3,label='$y=log(e^x)$')
plt.plot(x,y4,'o',lw=5,label='$y=(e^(logx))$')


plt.xlim([-2,2])
plt.ylim([-2,2])
plt.legend()

plt.show();


```
