```python
#Import Libraries
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

# Composite function

```python
#Using Numpy
def fx(x):
  return 2*x**2-4

def gx(x):
  return 7*np.abs(x)+3
```

```python
xx= np.linspace(-5,5,200)
fgx=fx(gx(xx))
gfx=gx(fx(xx))


#Plot
plt.plot(xx,fx(xx),lw=2,label='f(x)')
plt.plot(xx,gx(xx),lw=2,label='g(x)')
plt.plot(xx,fgx,lw=2,label='f(g(x))')
plt.plot(xx,gfx,lw=2,label='g(f(x))')

#plt.plot(xx[[0,-1]],[0,0],'k--',lw=.3)
#plt.plot([0,0],[-10,50],'k--',lw=.3)


plt.ylim([-10,50])
plt.xlim(xx[[-10,10]])
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.legend()
plt.show()

```

```python
# Another composite function

def fx(x):
  return np.sin(x)

def gx(x):
  return np.log(x)

def hx(x):
  return 2*x**2+5

xx=np.linspace(-100*np.pi,100*np.pi,1001)
f3=fx(gx(hx(xx)))



plt.plot(xx,f3,lw=2,label='f(g(h(x)))')
plt.xlabel='x'
plt.ylabel='y=f(x)'
plt.legend()
plt.show()
```

```python
#inverse function in sympy
sx=sym.symbols('x')
y= sym.symbols('y')

fun=2*sx+3

sym.solve(y-fun,sx)[0]
```
