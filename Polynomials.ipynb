```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

from IPython import display
display.set_matplotlib_formats('svg')
```

```python
#Numpy polynomials

xx=np.linspace(-5,5,100)

#random coeff
coefs=np.random.randn(4)

#construct polynomials

fname='$y='
y=np.zeros(len(xx))
print(coefs)
for i,c in enumerate(coefs):
  y+=c*xx**i
  fname += '+ '[int(c<0)] + f'{c:.2f}x^{i} '


plt.plot(xx,y,lw=2)
plt.title( fname+'$')
```

```python
#using Sympy
from sympy.abc import x
top=x**2-2*x
bottom=x**2-4
y=top/bottom

sym.plot(y,(x,-3,3),ylim=[-10,10],title=f'$y= {sym.latex(y)}$',ylabel=None,xlabel='x');
```

```python
#Sine wave with Polynomials
order=10

x=np.linspace(-2*np.pi,2*np.pi,100)
z=np.zeros(len(x))

for n in range (1, order+1):
  fun=(-1)**(n+1) * (x**(2*n-1))/np.math.factorial(2*n-1)
  plt.plot(x,fun,'--',lw=.8)

  z+=fun

plt.plot(x,z,'k',lw=2)

plt.plot(x[::5],np.sin(x[::5]))
plt.ylim=([-5,5])

```
