```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

from IPython.display import display,Math
```

```python
xx=np.linspace(-np.pi,2*np.pi,1234)

f=[0]*3


#function        labels
f[0]=np.sin(xx)   ,   'sin(x)'
f[1]=np.sin(xx)**2   , 'sin(x)**2'
f[2]=np.sin(xx**2)    ,'sin(x**2)'

for y,label in f:
  plt.plot(xx,y,label=label)

plt.legend()
plt.xlabel('Angle')
plt.ylabel("y = f(x)")
plt.show()

```

```python
#observe above figure red line is never below 0 as its square of sin(x)...so always +ve
#Green line when x increase sqare increases significantly means angle increses a lot hence increment in x brings more dense spike
```

```python
xx=np.linspace(-np.pi,2*np.pi,1234)
f=[0]*3

f[0]=np.sin(np.cos(xx))   , 'sin(cos(x))'
f[1]=np.cos(np.sin(xx))   , 'cos(sin(x)'
f[2]=np.cos(xx)           , 'cos(x)'


for y,label in f:
  plt.plot(xx,y,label=label)

plt.xlabel('angle')
plt.ylabel('y=f(x)')
plt.legend()
plt.show()
```
